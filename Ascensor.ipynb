```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Leer el archivo CSV (asegúrate de que el nombre coincida con tu archivo en "Files")
df = pd.read_csv('Pressure and velocity.csv', sep='\t')  # Cambia 'datos.csv' por tu nombre de archivo y ajusta 'sep' si es necesario

# 2. Mostrar las primeras filas y estadísticas
print("--- Primeras filas ---")
print(df.head())
print("\n--- Estadísticas ---")
print(df.describe())

# 3. Graficar (personaliza según tus columnas)
plt.figure(figsize=(15, 10))

# Gráfico 1: Altitud vs Tiempo
plt.subplot(2, 2, 1)
plt.plot(df['Time (s)'], df['Altitude (m)'], color='green', label='Altitud')
plt.title('Altitud vs Tiempo')
plt.xlabel('Time (s)')
plt.ylabel('Altitude (m)')
plt.legend()
plt.grid()

# Gráfico 2: Velocidad vs Tiempo (velocity)
plt.subplot(2, 2, 2)
plt.plot(df['Time (velocity) (s)'], df['Velocity (m/s)'], color='red', label='Velocidad')
plt.title('Velocidad vs Tiempo')
plt.xlabel('Time (velocity) (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.grid()

# Gráfico 3: Presión vs Altitud (scatter)
plt.subplot(2, 2, 3)
plt.scatter(df['Pressure (hPa)'], df['Altitude (m)'], color='blue', alpha=0.5, label='Presión')
plt.title('Presión vs Altitud')
plt.xlabel('Pressure (hPa)')
plt.ylabel('Altitude (m)')
plt.legend()
plt.grid()


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar el archivo CSV (asegúrate de que está en la pestaña 'Files' de Colab)
df = pd.read_csv('Acceleration.csv', sep='\t')  # Usamos sep='\t' porque tus datos están separados por tabulaciones

# 2. Verificar las columnas (opcional, para confirmar)
print("Columnas en el CSV:", df.columns.tolist())

# 3. Graficar
plt.figure(figsize=(12, 6))
plt.plot(df['Time (s)'], df['Acceleration (m/s²)'], color='purple', linewidth=1.5)
plt.title('Aceleración vs Tiempo', fontsize=14)
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Acceleration (m/s²)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
