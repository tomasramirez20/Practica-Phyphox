```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar el archivo CSV (usa sep='\t' si está separado por tabulaciones)
df = pd.read_csv('Raw Data.csv', sep='\t')  # Cambia 'datos.csv' por tu nombre de archivo

# 2. Extraer columnas (con los nombres exactos de tu CSV)
t = df['Time (s)']
x = df['Linear Acceleration x (m/s^2)']
y = df['Linear Acceleration y (m/s^2)']
z = df['Linear Acceleration z (m/s^2)']
a = df['Absolute acceleration (m/s^2)']

# 3. Configurar los subplots (4 gráficos verticales)
plt.figure(figsize=(10, 12))  # Tamaño de la figura

# Gráfico 1: Aceleración en X
plt.subplot(4, 1, 1)
plt.plot(t, x, color='blue', linewidth=1)
plt.title('Aceleración Lineal en X')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s²)')
plt.grid(True, linestyle='--', alpha=0.6)

# Gráfico 2: Aceleración en Y
plt.subplot(4, 1, 2)
plt.plot(t, y, color='red', linewidth=1)
plt.title('Aceleración Lineal en Y')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s²)')
plt.grid(True, linestyle='--', alpha=0.6)

# Gráfico 3: Aceleración en Z
plt.subplot(4, 1, 3)
plt.plot(t, z, color='green', linewidth=1)
plt.title('Aceleración Lineal en Z')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s²)')
plt.grid(True, linestyle='--', alpha=0.6)

# Gráfico 4: Aceleración Absoluta
plt.subplot(4, 1, 4)
plt.plot(t, a, color='purple', linewidth=1)
plt.title('Aceleración Absoluta')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s²)')
plt.grid(True, linestyle='--', alpha=0.6)

# Ajustar espacio entre subplots
plt.tight_layout()
plt.show()
```
